```python

student_files = glob.glob("stud*.csv")
student_files
```

```python
df_list = [pd.read_csv(df, sep='\t') for df in student_files]
students = pd.concat(df_list)
students = students.rename(columns={'Unnamed: 0': 'id'}).set_index('id')
students
```

```python
students.duplicated()
```

```python
students.value_counts()
```

```python
students = students.drop_duplicates()
students
```

```python
students.duplicated()
```

```python
students
```

```python
students['gender'] = students.gender_age.str[0]
students['age'] = students.gender_age.str[1:]
students = students.drop(columns='gender_age')
students
```

```python
str_split = pd.Series(students.full_name.str.split(' '))
str_split
```

```python
students['first_name'] = str_split.str.get(0)
students['second_name'] = str_split.str.get(1)
students
```

```python
students.dtypes
```

```python
students['fractions'] = students.fractions.replace('[\%,]', '', regex=True)
students
```

```python
students['fractions'] = pd.to_numeric(students.fractions)
students.dtypes
```

```python
grades = students.grade.str.replace('(\D)', '', regex=True)
grades
```

```python
students['grade'] = grades
students
```

```python
students.grade = pd.to_numeric(students.grade)
avg_grade = students.grade.mean()
avg_grade
```

```python
fractions_mean = students.fractions.mean().round(1)
fractions_mean
```

```python
students = students.fillna(value={'fractions': 0})
students
```
